```python
def add_values_in_dict(sample_dict, key, list_of_values):
    ''' Append multiple values to a key in 
        the given dictionary '''
    if key not in sample_dict:
        sample_dict[key] = list()
    sample_dict[key].extend(list_of_values)
    return sample_dict
```

```python
import warnings
from pathlib import Path
from Bio.PDB.PDBExceptions import PDBConstructionWarning
from Bio.PDB import PDBParser

# To get rid of those annoying warnings like 'WARNING: Chain B is discontinuous at line 4059.'
warnings.simplefilter('ignore', PDBConstructionWarning)

parser = PDBParser()

chainDict = dict()

if __name__ == "__main__":
    for pdb_filename in realpdbs:
        with open(pdb_filename, "rt") as file_handle:
            structure = parser.get_structure("?", file_handle)
    
        pdb_code = structure.header.get("idcode")
        
        for chain in structure.get_chains():
            chainId = []
            chainId.append(chain.id)
            chainDict = add_values_in_dict(chainDict, pdb_code, chainId)
```

```python
from Bio.PDB import PDBParser, PPBuilder

# Define a dictionary to map the three-letter codes to one-letter codes
aa_dict = {
    "ALA": "A",
    "ARG": "R",
    "ASN": "N",
    "ASP": "D",
    "CYS": "C",
    "GLN": "Q",
    "GLU": "E",
    "GLY": "G",
    "HIS": "H",
    "ILE": "I",
    "LEU": "L",
    "LYS": "K",
    "MET": "M",
    "PHE": "F",
    "PRO": "P",
    "SER": "S",
    "THR": "T",
    "TRP": "W",
    "TYR": "Y",
    "VAL": "V",
}

# Parse the PDB files and store sequences for each chain
parser = PDBParser()
chain_sequences = {}

for pdb_file in realpdbs:
    structure = parser.get_structure(pdb_file.strip('.pdb'), pdb_file)
    for chain in structure.get_chains():
        chain_id = chain.get_id()
        sequence = ""
        for residue in chain.get_residues():
            if residue.get_resname() != "HOH":
                residue_name = residue.get_resname()
                if residue_name in aa_dict:
                    sequence += aa_dict[residue_name]
                else:
                    sequence += "X"
        if (pdb_file, chain_id) in chain_sequences:
            chain_sequences[(pdb_file, chain_id)].append(sequence)
        else:
            chain_sequences[(pdb_file, chain_id)] = [sequence]

pdb_sequences = {}
for pdb_file in realpdbs:
    pdb_sequences[pdb_file] = {}
    structure = parser.get_structure(pdb_file.strip('.pdb'), pdb_file)
    for chain in structure.get_chains():
        chain_id = chain.get_id()
        if (pdb_file, chain_id) in chain_sequences:
            pdb_sequences[pdb_file][chain_id] = chain_sequences[(pdb_file, chain_id)]
        else:
            print(f"Chain {chain_id} not found in {pdb_file}.")
```
